```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '../ripe/ripe_durian_sensor_data_5.csv'
df = pd.read_csv(file_path)
heatmap_data = df.pivot_table(index="spot_y", columns="spot_x", values="sensor_reading", aggfunc="mean")
```

```python
sns.heatmap(heatmap_data, cmap="magma", annot=True, vmax=2,vmin=0)
```
